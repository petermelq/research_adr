```python
import xarray as xr

signal_betas = xr.open_dataarray('../data/processed/brit_underlying_betas_to_UKX_and_sector.nc')
hedge_betas = pd.read_csv('../data/processed/adr_EWU_betas.csv', index_col=0, parse_dates=True)
def get_predicted_returns(fut_intraday_ret, etf_intraday_ret, betas):
    # betas.index.name = 'date'
    # betas.columns.name = 'ticker'

    merged_df = pd.merge(fut_intraday_ret.stack().to_frame('future_returns'), 
                        etf_intraday_ret.stack().to_frame('etf_returns'),
                        on=['date','ticker'])
    merged_df = pd.merge(merged_df, betas.loc[:,:,'market_beta'].to_pandas().stack().to_frame('market_beta'), on=['date','ticker'])
    merged_df = pd.merge(merged_df, betas.loc[:,:,'sector_beta'].to_pandas().stack().to_frame('sector_beta'), on=['date','ticker'])
    merged_df['predicted_return'] = merged_df['market_beta'] * merged_df['future_returns'] + merged_df['sector_beta'].fillna(0) * merged_df['etf_returns']
    pred = merged_df['predicted_return'].unstack()

    return pred
```

```python
import pandas as pd
import numpy as np    

def attach_value_at_timestamp(df, value_col, timestamp_col, ref_col):
    output_col = 'output'
    df = df.sort_values(timestamp_col, kind='mergesort').reset_index(drop=True)

    ts = df[timestamp_col].to_numpy()
    mid = df[value_col].to_numpy()
    nopen = df[ref_col].to_numpy()

    # position of next_open in the timestamp array (first index where ts >= next_open)
    j = np.searchsorted(ts, nopen, side='left')

    out = np.full(len(df), np.nan, dtype=float)

    valid = (j < len(df)) & (ts[j] == nopen) & (j > np.arange(len(df)))
    out[valid] = mid[j[valid]]

    df[output_col] = out
    df['date'] = df[timestamp_col].dt.tz_localize(None).dt.normalize()
    df = df.set_index('date')
    
    return df[output_col]
```

```python
def attach_value_at_timestamp(df, value_col, timestamp_col, ref_col):
    df = df.copy()
    values = df.set_index(timestamp_col)[value_col].reindex(df[ref_col].unique()).to_frame()
    df[timestamp_col] = df[timestamp_col].dt.tz_localize(None).dt.normalize()
    df = df[[timestamp_col, ref_col]].drop_duplicates().merge(values, left_on=ref_col, right_index=True, how='left')
    return df.set_index(timestamp_col)[value_col]
```

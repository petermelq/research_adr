```python
import os
import shutil
from glob import glob

for path in glob('../data/raw/adrs/tcbbo/ticker=*'):
    _, ticker_dir = os.path.split(path)
    dest_dir = os.path.join('../data/raw/adrs/tcbbo/exchange=XNAS.BASIC/', ticker_dir)
    dest_parent = os.path.dirname(dest_dir)
    if os.path.exists(dest_dir):
        print(f"Directory {dest_dir} exists. Moving files into it.")
        for file in glob(os.path.join(path,'*')):
            print(f"Moving {file} to {dest_dir}")
            shutil.move(file, dest_dir)
    else:
        print(f"Moving {path} to {dest_parent}")
        shutil.move(path, dest_parent)
```

```python
import pandas as pd

df = pd.read_csv('../data/processed/market_etf_hedge_ratios.csv',index_col=0)
df
```

```python
import matplotlib.pyplot as plt
df.iloc[-350:].plot()
df.iloc[-350:].mean(axis=1).plot(linewidth=3, color='black', label='Average')

plt.xticks(rotation=45);
# putting legend outside of the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```

```python
import pandas as pd

df = pd.read_csv('../data/raw/all_adr_filtered.csv')
```

```python
from linux_xbbg import blp
import sys
sys.path.append('../src')
import utils

lookback = 60
cbday = utils.get_market_business_days('NYSE')
start_date = pd.Timestamp.today().normalize() - (lookback + 1) * cbday
end_date = pd.Timestamp.today().normalize() - cbday

turnover_df = blp.bdh(
    tickers=df['adr'].tolist(),
    flds=['PX_VOLUME'],
    start_date=start_date,
    end_date=end_date,
).droplevel(1,1).rename(columns=lambda x: x.replace(' US Equity', ''))
```

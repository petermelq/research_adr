```python
import pandas as pd
import pandas_market_calendars as mcal

adr_info = pd.read_csv('../data/raw/adr_info.csv')
exchanges = adr_info['exchange'].unique()
```

```python
max_by_ex = {}
for ex in exchanges:
    cal = mcal.get_calendar(ex)
    sched = cal.schedule(start_date='2025-01-01', end_date='2025-12-31')
    max_by_ex[ex] = sched['market_close'].dt.tz_convert('America/New_York').dt.time.max()

max_by_ex
```

```python
import numpy as np
import pandas as pd

adr_df = pd.read_csv('../data/processed/adr_mids_at_underlying_close_adjust_none.csv', index_col=0, parse_dates=True)
under_df = pd.read_csv('../data/processed/brit_underlying_close_to_usd_adr_midFromTick.csv', index_col=0, parse_dates=True)

#pd.read_csv('../data/processed/brit_underlying_close_to_usd_adr_PX_MID_adjust_all.csv', index_col=0, parse_dates=True)
#under_df = pd.read_csv('../data/processed/brit_underlying_close_to_usd_adr_PX_MID_adjust_none.csv', index_col=0, parse_dates=True)

holdings_filename = '../notebooks/brit_exchange_traded_efa_adrs_detailed.csv'
hold_df = pd.read_csv(holdings_filename)

adr_dict = hold_df[['id','adr']].set_index('id')['adr'].str.replace(' US Equity', '').to_dict()
under_df = under_df.rename(columns=adr_dict)

prem_df = ((adr_df.div(under_df)).dropna(axis=0, how='all')-1)*1e4
```

```python
import numpy as np
import pandas as pd

adr_df = pd.read_csv('../data/processed/adr_mids_at_underlying_close_adjust_all.csv', index_col=0, parse_dates=True)
under_df = pd.read_csv('../data/processed/brit_underlying_close_to_usd_adr_PX_MID_adjust_all.csv', index_col=0, parse_dates=True)

holdings_filename = '../notebooks/brit_exchange_traded_efa_adrs_detailed.csv'
hold_df = pd.read_csv(holdings_filename)

adr_dict = hold_df[['id','adr']].set_index('id')['adr'].str.replace(' US Equity', '').to_dict()
under_df = under_df.rename(columns=adr_dict)

adj_prem_df = ((adr_df.div(under_df)).dropna(axis=0, how='all')-1)*1e4
```

```python
hold_df[['id','adr']].to_clipboard(index=False, header=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

for ticker in prem_df.columns:
    plt.figure()
    prem_df[ticker].plot(label='Unadjusted')
    adj_prem_df[ticker].plot(label='Adjusted')
    plt.title(f'{ticker} (mean: {prem_df[ticker].iloc[-250:].mean():.1f})')
    plt.legend()
```

```python
for ticker in prem_df.columns:
    plt.figure()
    series = adj_prem_df[ticker].iloc[-200:]
    series.plot()
    series.ewm(halflife=5).mean().plot()
    plt.title(f'{ticker} (mean: {series.mean():.1f})')
```
